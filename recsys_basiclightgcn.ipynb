```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
models = ['basicTransformer_Rerank', 'basicDLCM', 'basicLightGCN', 'basicNGCF']
ndcg_1 = [0.58671, 0.51917, 0.24005, 0.24025]
ndcg_3 = [0.61540, 0.54759, 0.26856, 0.27150]
ndcg_5 = [0.68533, 0.62750, 0.34336, 0.34758]
ndcg_7 = [0.71374, 0.67128, 0.42439, 0.42743]

# 그래프 스타일 설정
plt.figure(figsize=(10, 6))
plt.grid(True, linestyle='--', alpha=0.7)

# 각 @K별로 선 그리기
plt.plot(models, ndcg_1, 'b-o', label='@1')
plt.plot(models, ndcg_3, 'y-o', label='@3')
plt.plot(models, ndcg_5, 'g-o', label='@5')
plt.plot(models, ndcg_7, 'r-o', label='@7')

# 그래프 꾸미기
plt.xlabel('Models')
plt.ylabel('NDCG')
plt.title('NDCG Comparison across Models')
plt.legend()
plt.xticks(rotation=45)

# 그래프 표시
plt.tight_layout()
plt.show()

# 각 포인트에 값 표시하기
for i, model in enumerate(models):
    plt.text(i, ndcg_1[i], f'{ndcg_1[i]:.5f}', color='blue', fontsize=9, ha='center', va='bottom')
    plt.text(i, ndcg_3[i], f'{ndcg_3[i]:.5f}', color='yellow', fontsize=9, ha='center', va='bottom')
    plt.text(i, ndcg_5[i], f'{ndcg_5[i]:.5f}', color='green', fontsize=9, ha='center', va='bottom')
    plt.text(i, ndcg_7[i], f'{ndcg_7[i]:.5f}', color='red', fontsize=9, ha='center', va='bottom')
```
